Import Statements

```
# All import statements
import pandas as pd
import numpy as np
```

Read From CSV File


```
# Read all CSV Files
train_df = pd.read_csv('https://raw.githubusercontent.com/rohithreddyvennam/AssignmentsFiles/master/train.csv')
test_df = pd.read_csv('https://raw.githubusercontent.com/rohithreddyvennam/AssignmentsFiles/master/test.csv')
submission_df = pd.read_csv('https://raw.githubusercontent.com/rohithreddyvennam/AssignmentsFiles/master/sampleSubmission.csv')
```

Set Sample Train, Test, and output tables

```
temp_index = train_df['Id']>=5000
temp_df = train_df[(temp_index)]

train_sample_index= np.logical_not(temp_index)
train_sample_df = train_df[(train_sample_index)]

test_sample_index = (temp_df['Id']+1)%5 != 0
test_sample_df = temp_df[(test_sample_index)]

result_sample_index = np.logical_not(test_sample_index) 
result_sample_df = temp_df[(result_sample_index)]

#print(*temp, sep = "\n")
```

```

train_sample_df = train_df
test_sample_df = test_df


#print(*temp, sep = "\n")
```

Trim Id's

```
train_sample_trim_df = train_sample_df.drop(['Id'], axis = 1)
test_sample_trim_df = test_sample_df.drop(['Id'], axis = 1)
result_sample_trim_df = result_sample_df.drop(['Id'], axis = 1)
```

```
pd.set_option("display.precision", 24)

# changing object type to complex type
df_train_complex = train_sample_trim_df.apply(lambda col: col.apply(lambda val: complex(val.strip('()'))))
df_real = df_train_complex.apply(np.real)
df_imag = df_train_complex.apply(np.imag)

# changing column names
for ind in range(24):
  df_real.rename(index=str, columns={str(ind): (str(ind)+"r")}, inplace=True)
  df_imag.rename(index=str, columns={str(ind): (str(ind)+"i")}, inplace=True)

df_train = pd.concat([df_real, df_imag], axis=1, join='inner')
#df1.append(df2)
```

```
pd.set_option("display.precision", 24)

# changing object type to complex type
df_test_complex = test_sample_trim_df.apply(lambda col: col.apply(lambda val: complex(val.strip('()'))))
df_real = df_test_complex.apply(np.real)
df_imag = df_test_complex.apply(np.imag)

# changing column names
for ind in range(24):
  df_real.rename(index=str, columns={str(ind): (str(ind)+"r")}, inplace=True)
  df_imag.rename(index=str, columns={str(ind): (str(ind)+"i")}, inplace=True)

df_test = pd.concat([df_real, df_imag], axis=1, join='inner')
df_test
#df1.append(df2)
```

```
pd.set_option("display.precision", 24)

# changing object type to complex type
df_result_complex = result_sample_trim_df.apply(lambda col: col.apply(lambda val: complex(val.strip('()'))))
df_real = df_result_complex.apply(np.real)
df_imag = df_result_complex.apply(np.imag)

# changing column names
for ind in range(24):
  df_real.rename(index=str, columns={str(ind): (str(ind)+"r")}, inplace=True)
  df_imag.rename(index=str, columns={str(ind): (str(ind)+"i")}, inplace=True)

df_result = pd.concat([df_real, df_imag], axis=1, join='inner')
df_result
#df1.append(df2)
```

```
df_submit = df_result 
for col in df_submit.columns:
    df_submit[col].values[:] = 0

df_submit.to_csv('submit_sample.csv')
```
